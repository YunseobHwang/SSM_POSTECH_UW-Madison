````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import itertools

import sys
sys.path.append('/mnt/disk1/yunseob/Pytorch/SSM/0_modules')
from SSM_utils import GBdataLoad, B6dataLoad
from SSM_DL_evl import SSM_Model

from sklearn.metrics import roc_curve, auc
````

````python
model_dir = './model/'
CM_save_dir = './results/1_confusion_mat/'
````

````python
data_dir = '/mnt/disk1/yunseob/Pytorch/SSM/1_Semblex/05_10p_overlap/SemblexData/wt_4040'

GB_test_X, GB_test_Y = GBdataLoad(data_dir, ch = [0, 1], data_type = '4_test_data')
B6_test_X, B6_test_Y = B6dataLoad(data_dir, ch = [0, 1], data_type = '4_test_data')
````

````python
GB_train_X, GB_train_Y = GBdataLoad(data_dir, ch = [0, 1], data_type = '3_train_data')
B6_train_X, B6_train_Y = B6dataLoad(data_dir, ch = [0, 1], data_type = '3_train_data')

GB_train_X, GB_train_Y = np.concatenate(GB_train_X), np.concatenate(GB_train_Y)
B6_train_X, B6_train_Y = np.concatenate(B6_train_X), np.concatenate(B6_train_Y)
````

# 1. Fault diagnosis

````python
class_name = ['Good', 'Bad']
````

````python
GB_10p_AB_model = SSM_Model(GB_train_X, GB_train_Y, n_ch = 2, n_cls = 2, class_name = class_name, 
                            model_dir = model_dir, model_name = 'GB_10p_AB', batch_size = 100, GPU_idx = 6)

GB_10p_AB_model.Evaluation()
````

````python
GB_10p_AB_model = SSM_Model(GB_test_X, GB_test_Y, n_ch = 2, n_cls = 2, class_name = class_name, 
                            model_dir = model_dir, model_name = 'GB_10p_AB', batch_size = 200, GPU_idx = 3)

GB_10p_AB_model.Evaluation()
GB_10p_AB_model.ConfusionMatrix(CM_save_dir, x_angle = 0)
````

````python
GB_10p_A_model = SSM_Model(np.expand_dims(GB_train_X[:,0,:,:], axis = 1), GB_train_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_A', batch_size = 200, GPU_idx = 3)

GB_10p_A_model.Evaluation()
````

````python
GB_10p_A_model = SSM_Model(np.expand_dims(GB_test_X[:,0,:,:], axis = 1), GB_test_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_A', batch_size = 200, GPU_idx = 3)

GB_10p_A_model.Evaluation()
````

````python
GB_10p_A_model = SSM_Model(np.expand_dims(GB_test_X[:,0,:,:], axis = 1), GB_test_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_A', batch_size = 200, GPU_idx = 3)

GB_10p_A_model.Evaluation()
GB_10p_A_model.ConfusionMatrix(CM_save_dir, x_angle = 0)
````

````python
GB_10p_B_model = SSM_Model(np.expand_dims(GB_train_X[:,1,:,:], axis = 1), GB_train_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_B', batch_size = 200, GPU_idx = 3)

GB_10p_B_model.Evaluation()
````

````python
GB_10p_B_model = SSM_Model(np.expand_dims(GB_test_X[:,1,:,:], axis = 1), GB_test_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_B', batch_size = 200, GPU_idx = 3)

GB_10p_B_model.Evaluation()
````

````python
GB_10p_B_model = SSM_Model(np.expand_dims(GB_test_X[:,1,:,:], axis = 1), GB_test_Y, n_ch = 1, n_cls = 2, 
                           class_name = class_name, model_dir = model_dir, model_name = 'GB_10p_B', batch_size = 200, GPU_idx = 3)

GB_10p_B_model.Evaluation()
GB_10p_B_model.ConfusionMatrix(CM_save_dir, x_angle = 0)
````

````python
df = pd.DataFrame(columns = ['Accuracy', 'Precision', 'Recall', 'F1 score'])

CMs = [GB_10p_A_model.cm, GB_10p_B_model.cm, GB_10p_AB_model.cm]
model_names = [GB_10p_A_model.model_name, GB_10p_B_model.model_name, GB_10p_AB_model.model_name]

for cm, model_name in zip(CMs, model_names):
    accr = round((cm[0, 0] + cm[1, 1]) / np.sum(cm), 4)
    precision = round(cm[1, 1] / (cm[0, 1] + cm[1, 1]), 4)
    recall = round(cm[1, 1] / (cm[1, 0] + cm[1, 1]), 4)
    f1_score = round(2*precision*recall/(precision + recall), 4)
    df.loc[model_name] = [accr, precision, recall, f1_score]

df.to_csv('./results/GB_10p_cls_performance.csv',encoding='utf-8-sig', index = True)    
    
df
````

````python
def ROCplot(test_Y, model, color, model_name = ''):
    fpr, tpr, _ = roc_curve(test_Y, model.score[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, color = color, lw=1.2, label='{} (AUC = {:.4f})'.format(model_name, roc_auc))
    # plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

fig = plt.figure(figsize = (10, 8))

ROCplot(GB_test_Y, GB_10p_A_model, 'g', 'A1')
ROCplot(GB_test_Y, GB_10p_B_model, 'c', 'A2')
ROCplot(GB_test_Y, GB_10p_AB_model, 'r', 'A1 & A2')

plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.01])
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('False Positive Rate', fontsize = 16)
plt.ylabel('True Positive Rate', fontsize = 16)
# plt.title('Receiver operating characteristic')
plt.legend(loc="lower right", fontsize = 14)
plt.show()

fig.savefig('./results/GB_10p_ROC_curves.png', bbox_inches='tight')
````

````python
model_names = [B6_10p_A_model.model_name, B6_10p_B_model.model_name, B6_10p_AB_model.model_name]
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 score']
tuples = [i for i in itertools.product(model_names, metrics)]
col = pd.MultiIndex.from_tuples(tuples)
df = pd.DataFrame(columns = col)

CMs = [B6_10p_A_model.cm, B6_10p_B_model.cm, B6_10p_AB_model.cm]
class_names = ['Heavy Oil', 'Die Punch', 'Scrapped Wire', 'Die Chip', 'Die Internal', 'PIN']
for c_n, cm in zip(np.arange(0, 16, 4), CMs):
    accr = round(np.trace(cm) / np.sum(cm), 4)
    for i, cls_name in zip(range(len(class_names)), class_names):
        precision = round(cm[i, i] / np.sum(cm[:, i]), 4)
        recall = round(cm[i, i] / np.sum(cm[i, :]), 4)
        f1_score = round(2*precision*recall/(precision + recall), 4)
        df.loc[cls_name, c_n:c_n+4] = [accr, precision, recall, f1_score]
        
df.to_csv('./results/B6_10p_cls_performance.csv',encoding='utf-8-sig', index = True)    

df
````
