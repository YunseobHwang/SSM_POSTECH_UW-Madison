```python
BAD_cls = {0: 'OIL', 1: 'PUNCH', 2: 'SCRAPPED', 3: 'DIE_CHIP', 4: 'DIE_INTERNAL', 5: 'PIN'}

B_train_X1, B_train_X2, B_train_X3, B_train_X4, B_train_X5, B_train_X6 = [], [], [], [], [], []
for cls, data_bunch in data.items():
    for data_i in data_bunch:
        if cls == 'BAD':
            if BAD_cls[0] in data_i['1_file']: B_train_X1.append(data_i['3_train_data'][:,:,:, 0])
            if BAD_cls[1] in data_i['1_file']: B_train_X2.append(data_i['3_train_data'][:,:,:, 0])
            if BAD_cls[2] in data_i['1_file']: B_train_X3.append(data_i['3_train_data'][:,:,:, 0])
            if BAD_cls[3] in data_i['1_file']: B_train_X4.append(data_i['3_train_data'][:,:,:, 0])
            if BAD_cls[4] in data_i['1_file']: B_train_X5.append(data_i['3_train_data'][:,:,:, 0])
            if BAD_cls[5] in data_i['1_file']: B_train_X6.append(data_i['3_train_data'][:,:,:, 0])
                
def concat(data):
    return np.expand_dims(np.concatenate(data), axis = 3)

B_train_X1, B_train_X2, B_train_X3, B_train_X4, B_train_X5, B_train_X6 = (concat(B_train_X1), concat(B_train_X2), 
                                                                    concat(B_train_X3), concat(B_train_X4), 
                                                                    concat(B_train_X5), concat(B_train_X6))

B_train_X1.shape, B_train_X2.shape, B_train_X3.shape, B_train_X4.shape, B_train_X5.shape, B_train_X6.shape
```

```python
def input_for_Pytorch(img):
    """
    (sample #, height, width, channel) -> (sample #, channel, height, width)
    """
    img = np.transpose(img, (0, 3, 1, 2))
    return img

B_train_X1, B_train_X2, B_train_X3, B_train_X4, B_train_X5, B_train_X6 = (input_for_Pytorch(B_train_X1), input_for_Pytorch(B_train_X2), 
                                                                    input_for_Pytorch(B_train_X3), input_for_Pytorch(B_train_X4), 
                                                                    input_for_Pytorch(B_train_X5), input_for_Pytorch(B_train_X6))

B_train_X1.shape, B_train_X2.shape, B_train_X3.shape, B_train_X4.shape, B_train_X5.shape, B_train_X6.shape
```

```python
def CreateLabel(data, cls):
    label = cls*np.ones([data.shape[0]])
    return label

B_train_Y1, B_train_Y2, B_train_Y3, B_train_Y4, B_train_Y5, B_train_Y6 = (CreateLabel(B_train_X1, 0), CreateLabel(B_train_X2, 1), 
                                                                    CreateLabel(B_train_X3, 2), CreateLabel(B_train_X4, 3), 
                                                                    CreateLabel(B_train_X5, 4), CreateLabel(B_train_X6, 5))

B_train_Y1.shape, B_train_Y2.shape, B_train_Y3.shape, B_train_Y4.shape, B_train_Y5.shape, B_train_Y6.shape
```

```python
def train_valid_split(data, label, rate_t_v = 0.9):
    data_num = len(data)
    train_idx = np.random.choice(data_num, int(rate_t_v*data_num), replace = False)
    valid_idx = np.setdiff1d(np.arange(data_num), train_idx)
    return data[train_idx], label[train_idx], data[valid_idx], label[valid_idx]

B_train_X1, B_train_Y1, B_valid_X1, B_valid_Y1 = train_valid_split(B_train_X1, B_train_Y1)
B_train_X2, B_train_Y2, B_valid_X2, B_valid_Y2 = train_valid_split(B_train_X2, B_train_Y2)
B_train_X3, B_train_Y3, B_valid_X3, B_valid_Y3 = train_valid_split(B_train_X3, B_train_Y3)
B_train_X4, B_train_Y4, B_valid_X4, B_valid_Y4 = train_valid_split(B_train_X4, B_train_Y4)
B_train_X5, B_train_Y5, B_valid_X5, B_valid_Y5 = train_valid_split(B_train_X5, B_train_Y5)
B_train_X6, B_train_Y6, B_valid_X6, B_valid_Y6 = train_valid_split(B_train_X6, B_train_Y6)

print("Training:", B_train_Y1.shape, B_train_Y2.shape, B_train_Y3.shape, B_train_Y4.shape, B_train_Y5.shape, B_train_Y6.shape)
print("Validation:", B_valid_Y1.shape, B_valid_Y2.shape, B_valid_Y3.shape, B_valid_Y4.shape, B_valid_Y5.shape, B_valid_Y6.shape)
```

# Model

```python
class ConvNet(nn.Module):
    def __init__(self):
        super().__init__()
        # RGB세개 1채널, 20개 특징 추출, filter 크기, stride 1
        self.conv1_1 = nn.Conv2d(1, 64, 3, 1, padding=1)
        self.conv1_2 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv1_bn = nn.BatchNorm2d(64)
        
        self.conv2_1 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv2_2 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv2_bn = nn.BatchNorm2d(64)
        
        self.conv3_1 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv3_2 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv3_bn = nn.BatchNorm2d(64)
        
        self.conv4_1 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv4_2 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv4_bn = nn.BatchNorm2d(64)
        
        self.dense1 = nn.Linear(2*2*64, 128)
        # 0.5 가 권장 할 만하대
        self.dropout1 = nn.Dropout(0.5)
        self.dense2 = nn.Linear(128, 6)  
        
    def forward(self, x):
        x = F.relu(self.conv1_1(x))
        x = F.relu(self.conv1_bn(self.conv1_2(x)))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv2_1(x))
        x = F.relu(self.conv2_bn(self.conv2_2(x)))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv3_1(x))
        x = F.relu(self.conv3_bn(self.conv3_2(x)))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv4_1(x))
        x = F.relu(self.conv4_bn(self.conv4_2(x)))
        x = F.max_pool2d(x, 2, 2)
        # flatten
        x = x.view(-1, 2*2*64)
        x = F.relu(self.dense1(x))
        x = self.dropout1(x)
        x = self.dense2(x)
        return x
```

```python
model = ConvNet()
model = model.cuda()
# model = model.to(device)
# if device == 'cuda':
#     net = torch.nn.DataParallel(net)
#     cudnn.benchmark = True
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.00001)

summary(model, (1, 40, 40))
```

# Training

```python
def RandomMinibatch(data, label, n_batch = 100):
    if n_batch <= len(data):
        batch_idx = np.random.choice(len(data), n_batch, replace = False)
    else:
        batch_idx = np.random.choice(len(data), len(data), replace = False)
    return data[batch_idx], label[batch_idx]

def shuffle(x, y):
    """
    random shuffle of two paired data -> x, y = shuffle(x, y)
    but, available of one data -> x = shuffle(x, None)
    """
    idx = np.arange(len(x))
    np.random.shuffle(idx)
    if type(x) == type(y):
        return x[idx], y[idx] 
    else:
        return x[idx]
```

```python
loss_history, accr_history = [], []
val_loss_history, val_accr_history = [], []

iter_i = 0

while True:
    iter_i += 1
    
    B_train_x1, B_train_y1 = RandomMinibatch(B_train_X1, B_train_Y1, 30)
    B_train_x2, B_train_y2 = RandomMinibatch(B_train_X2, B_train_Y2, 30)
    B_train_x3, B_train_y3 = RandomMinibatch(B_train_X3, B_train_Y3, 30)
    B_train_x4, B_train_y4 = RandomMinibatch(B_train_X4, B_train_Y4, 30)
    B_train_x5, B_train_y5 = RandomMinibatch(B_train_X5, B_train_Y5, 30)
    B_train_x6, B_train_y6 = RandomMinibatch(B_train_X6, B_train_Y6, 30)
    train_x = np.vstack([B_train_x1, B_train_x2, B_train_x3, B_train_x4, B_train_x5, B_train_x6])
    train_y = np.hstack([B_train_y1, B_train_y2, B_train_y3, B_train_y4, B_train_y5, B_train_y6])

    train_x, train_y = shuffle(train_x, train_y)
    
    train_x, train_y = torch.tensor(train_x, device=device).float(), torch.tensor(train_y, device=device).long()
    
    output = model(train_x)
    loss = criterion(output, train_y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if iter_i % 10 == 0:
        
        _, pred = torch.max(output, 1)
        
        loss_history.append(loss.tolist())
        accr_history.append((torch.sum(pred == train_y.data).tolist() / len(train_y)))
        
        with torch.no_grad():
            
            B_valid_x1, B_valid_y1 = RandomMinibatch(B_valid_X1, B_valid_Y1, 30)
            B_valid_x2, B_valid_y2 = RandomMinibatch(B_valid_X2, B_valid_Y2, 30)
            B_valid_x3, B_valid_y3 = RandomMinibatch(B_valid_X3, B_valid_Y3, 30)
            B_valid_x4, B_valid_y4 = RandomMinibatch(B_valid_X4, B_valid_Y4, 30)
            B_valid_x5, B_valid_y5 = RandomMinibatch(B_valid_X5, B_valid_Y5, 30)
            B_valid_x6, B_valid_y6 = RandomMinibatch(B_valid_X6, B_valid_Y6, 30)
            valid_x = np.vstack([B_valid_x1, B_valid_x2, B_valid_x3, B_valid_x4, B_valid_x5, B_valid_x6])
            valid_y = np.hstack([B_valid_y1, B_valid_y2, B_valid_y3, B_valid_y4, B_valid_y5, B_valid_y6])

            valid_x, valid_y = torch.tensor(valid_x, device=device).float(), torch.tensor(valid_y, device=device).long()

            valid_output = model(valid_x)
            valid_loss = criterion(valid_output, valid_y)
            
            _, valid_pred = torch.max(valid_output, 1)
        
            val_loss_history.append(valid_loss.tolist())
            val_accr_history.append((torch.sum(valid_pred == valid_y.data).tolist() / len(valid_y)))
        
        print("{:05d} | train_loss: {:.5f}, train_accr: {:.2f} | val_loss: {:.5f}, val_accr: {:.2f}".format(iter_i, 
                                                                                                            loss_history[-1], 
                                                                                                            accr_history[-1], 
                                                                                                            val_loss_history[-1], 
                                                                                                            val_accr_history[-1]))
        
        if iter_i >= 100:  
            if np.mean(val_loss_history[-5 :]) >= np.mean(val_loss_history[-10:]):
                print('')
                print('Early stopping!!')
                print('..val_loss (avg. within 5): {:.5f} >= val_loss (avg. within 10) {:.5f}'.format(np.mean(val_loss_history[-5 :]),
                                                                                                  np.mean(val_loss_history[-10:])))
                
                now = datetime.datetime.now()
                nowDatetime = now.strftime('%y%m%d%H%M')
                model_name = 'B6_sem_A_{}_{:05d}_loss_{:.6f}_val_loss_{:.6f}.pt'.format(nowDatetime, iter_i, 
                                                                                     np.mean(loss_history[-3:]),
                                                                                     np.mean(val_loss_history[-3:]))
                torch.save(model.state_dict(), './model/' + model_name)
                break   
```

```python
fig = plt.figure(figsize = (15,17))
plt.suptitle('Training History', y = 0.92, fontsize = 20)

# x_axis = range(1, 10*len(accr_history)+1, 10)
x_axis = np.arange(10, 10*len(accr_history)+1, 10)

plt.subplot(2, 1, 1)
plt.plot(x_axis, accr_history, 'b-', label = 'Training Accuracy')
plt.plot(x_axis, val_accr_history, 'r-', label = 'Validation Accuracy')
plt.xlabel('Iteration', fontsize = 15)
plt.ylabel('Accuracy', fontsize = 15)
plt.legend(fontsize = 10)
plt.subplot(2, 1, 2)
plt.plot(x_axis, loss_history, 'b-', label = 'Training Loss')
plt.plot(x_axis, val_loss_history, 'r-', label = 'Validation Loss')
plt.xlabel('Iteration', fontsize = 15)
plt.ylabel('Loss', fontsize = 15)
# plt.yticks(np.arange(0, 0.25, step=0.025))
plt.legend(fontsize = 12)
plt.show()

# if save == True:
#     fig.savefig(hist_path + filename)
#     plt.close(fig)
```

```python
def plotCM(cm, class_name, value_size, label_size, mode):

    plt.imshow(cm, interpolation = 'nearest', cmap = plt.cm.Blues)
    thresh = cm.max()/2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if mode == 'percent':
            value = np.round(cm[i, j]/(np.sum(cm, 1)[i]), 3)
        if mode == 'num':
            value = cm[i, j]
        plt.text(j, i, value,
                 fontsize = value_size,
                 horizontalalignment = 'center',
                 color = 'white' if cm[i, j] > thresh else 'black')
    plt.ylabel('True label', fontsize = label_size)
    plt.xlabel('Predicted', fontsize = label_size)
    plt.xticks([i for i in range(len(class_name))], class_name, rotation=270, fontsize = label_size)
    plt.yticks([i for i in range(len(class_name))], class_name, rotation=0, fontsize = label_size)
```
