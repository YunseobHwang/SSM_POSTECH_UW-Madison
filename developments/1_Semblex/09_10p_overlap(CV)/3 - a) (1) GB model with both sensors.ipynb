```python
import sys
sys.path.append('/mnt/disk1/yunseob/Pytorch/2_SSM/0_modules')
from SSM_utils import GBdataLoad
from SSM_DL_training import CNN_Training
```

```python
data_dir = '/mnt/disk1/yunseob/Pytorch/2_SSM/1_Semblex/05_10p_overlap/SemblexData/wt_4040'

GB_Xs, GB_Ys = GBdataLoad(data_dir, ch = [0, 1], data_type = '2_data')
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5, shuffle = True, random_state = 44)

for train_idx, test_idx in kf.split(GB_Xs[0]):
    print('test:', test_idx)
```

```python
def CV_train_idx(X, n_splits = 5, shuffle = True, random_state = 44):
    kf = KFold(n_splits, shuffle, random_state)
    train_idx = []
    for t_idx, _ in kf.split(X):
        train_idx.append(t_idx)
    return train_idx

G_train_idx = CV_train_idx(GB_Xs[0])
B_train_idx = CV_train_idx(GB_Xs[1])
```

```python
LRs = [0.01, 0.001, 0.0001]

for lr in LRs:
    model_spec = 1
    for i, g_train_idx, b_train_idx in zip(range(5), G_train_idx, B_train_idx):
        GB_Xs_cv = [GB_Xs[0][g_train_idx], GB_Xs[1][b_train_idx]]
        GB_Ys_cv = [GB_Ys[0][g_train_idx], GB_Ys[1][b_train_idx]]

        CT = CNN_Training(GB_Xs_cv, GB_Ys_cv, n_ch = 2, n_cls = 2, 
                          lr = 0.01, n_batch_per_cls = 16, n_epoch = 500, n_patience = 30, 
                          model_name = '{:02d}_GB_10p_AB'.format(i+1), GPU_idx = 3)
        if i != 0:
            model_spec = 0
        CT.Run(model_spec = model_spec, verbose = 2, model_dir = './model/')
```
