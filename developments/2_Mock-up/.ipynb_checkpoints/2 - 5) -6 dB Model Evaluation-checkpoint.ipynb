```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import os

import sys
sys.path.append('/mnt/disk1/yunseob/Pytorch/SSM/0_modules')
from SSM_utils import Reshape4torch, GenerateLabel
from SSM_DL_evl import SSM_Model

from sklearn.metrics import roc_curve, auc
```

# 1. Load Testing Data

```python
data_dir = '/mnt/disk2/data/private_data/SSM/2_Mockup/1_AE_Accel_Data/2_NOISE_GN/-6db'

def GBdataLoad2(data_dir, phase = 'train', data = 'AE'):
    for i in os.listdir(data_dir):
        if 'good' in i and phase in i and data in i:
            G_X = np.load(os.path.join(data_dir, i))
        elif 'bad' in i and phase in i and data in i:
            B_X = np.load(os.path.join(data_dir, i))
            
    def add_ch(img):
        """
        (sample #, height, width,) -> (sample #, height, width, channel)
        """
        return np.expand_dims(img, axis = -1)
    
    G_X, B_X = add_ch(G_X), add_ch(B_X)
    G_X, B_X = Reshape4torch(G_X), Reshape4torch(B_X)

    return G_X, B_X

AE_G_X, AE_B_X = GBdataLoad2(data_dir, phase = 'test', data = 'AE')
Accel_G_X, Accel_B_X = GBdataLoad2(data_dir, phase = 'test', data = 'accel')
A2_G_X = np.concatenate((AE_G_X, Accel_G_X), axis = 1)
A2_B_X = np.concatenate((AE_B_X, Accel_B_X), axis = 1)
G_Y, B_Y = GenerateLabel(A2_G_X, 0), GenerateLabel(A2_B_X, 1)

AE_X = np.concatenate((AE_G_X, AE_B_X), axis = 0)
Accel_X = np.concatenate((Accel_G_X, Accel_B_X), axis = 0)
A2_X = np.concatenate((A2_G_X, A2_B_X), axis = 0)
Y = np.concatenate((G_Y, B_Y), axis = 0)

print('AE data', AE_G_X.shape, AE_B_X.shape,  '=', AE_X.shape, '\n')
print('Accel data', Accel_G_X.shape, Accel_B_X.shape,  '=', Accel_X.shape, '\n')
print('A2 data:', A2_G_X.shape, A2_B_X.shape, '=', A2_X.shape)
```

# 2. Evaluation

```python
model_dir = '3 ./model/'
CM_save_dir = './results/1_confusion_mat/'
```

```python
class_name = ['Good', 'Bad']

GB_GN_m6dB_AE_model = SSM_Model(AE_X, Y, n_ch = 1, n_cls = 2, class_name = class_name, 
                            model_dir = model_dir, model_name = 'GB_GN_-6dB_AE', GPU_idx = 3)

GB_GN_m6dB_AE_model.Evaluation(CM_save_dir, x_angle = 0)
```

```python
GB_GN_m6dB_Accel_model = SSM_Model(Accel_X, Y, n_ch = 1, n_cls = 2, class_name = class_name, 
                            model_dir = model_dir, model_name = 'GB_GN_-6dB_Accel', GPU_idx = 3)

GB_GN_m6dB_Accel_model.Evaluation(CM_save_dir, x_angle = 0)
```

```python
GB_GN_m6dB_A2_model = SSM_Model(A2_X, Y, n_ch = 2, n_cls = 2, class_name = class_name, 
                            model_dir = model_dir, model_name = 'GB_GN_-6dB_A2', GPU_idx = 3)

GB_GN_m6dB_A2_model.Evaluation(CM_save_dir, x_angle = 0)
```
