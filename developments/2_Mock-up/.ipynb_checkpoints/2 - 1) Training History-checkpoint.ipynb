# 2. Plot and Save

```python
save_dir = './results/0_train_hist/graph/'

def HISTsave(hist, val_hist, model_name = 'AE_GB_model', value_name = 'loss', y_value = 'Loss'):
    
    n_iter = len(hist)
    interval = int(round(n_iter/5, -1))
    fig = plt.figure(figsize = (12,10))

    plt.plot(hist, 'b', label = 'TRAINING')
    plt.plot(val_hist, 'r', label = 'VALIDATION')
    plt.xlabel('Iteration', fontsize = 16)
    plt.ylabel(y_value, fontsize = 16)
    plt.xticks(list(np.arange(0, n_iter+1, interval)), list(10*np.arange(0, n_iter+1, interval)), fontsize = 13)
    plt.yticks(fontsize = 13)
    plt.legend(fontsize = 14)
    plt.grid('on')
    plt.show()
    
    file_name = '{}_{}_hist.png'.format(model_name, value_name)
    fig.savefig(save_dir + file_name, bbox_inches='tight')
    plt.close(fig) 
```
